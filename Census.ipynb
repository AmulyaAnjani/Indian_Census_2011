## trying to plot the different checkpoints for the datasets by comparing different columsn in the table 

```python
l1 = len(d1)

bihar_districts = [d1.iloc[index][2] for index in range(l1)]

bihar_pop = [d1.iloc[i][30] for i in range(l1)]

l2 = len(d2)

tamil_districts = [d2.iloc[ind][2] for ind in range(l2)]    

tamil_population = [d2.iloc[u][30] for u in range(l2)]
```

```python
plt.xticks(range(l1)+range(l2),bihar_districts+tamil_districts)
districts = bihar_pop+tamil_population
plt.scatter(range(l1)+range(l2),districts,color=["blue" if dit in bihar_pop else "red" for dit in districts])
```

```python
plt.show()
```
